```python
#user-id별 rating 횟수 & rating 평균
df_rating_Guser = df_rating.groupby('User-ID')
df_rating_Guser = df_rating_Guser.agg({'ISBN' : 'count', 'Book-Rating' : 'mean'})
df_rating_Guser = df_rating_Guser.reset_index()
df_rating_Guser.columns = ['User-ID', 'RatingCount', 'RatingMean']

df_rating_Guser.describe()
```

```python
df_rating_Guser.corr()
```

```python
df_rating2 = df_rating[df_rating['Book-Rating'] > 0]

#user-id별 rating 횟수 & rating 평균
df_rating_Guser = df_rating2.groupby('User-ID')
df_rating_Guser = df_rating_Guser.agg({'ISBN' : 'count', 'Book-Rating' : 'mean'})
df_rating_Guser = df_rating_Guser.reset_index()
df_rating_Guser.columns = ['User-ID', 'RatingCount', 'RatingMean']

df_rating_Guser.describe()
```
